# overfitting graph

```python
from sklearn.datasets import make_classification
```

```python
X,y= make_classification(n_samples=10000,n_features=20, n_informative=5,
    n_redundant=15,random_state=3)
```

```python
X.shape
```

```python
y.shape
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
train = []
test = []

# Loop through 20 iterations
for i in range(1, 21):
    model = DecisionTreeClassifier(max_depth=i)
    model.fit(X_train, y_train)
    train_yhat = model.predict(X_train)
    train_acc = accuracy_score(y_train, train_yhat)
    train.append(train_acc)  # Store training accuracy
    test_yhat = model.predict(X_test)
    test_acc = accuracy_score(y_test, test_yhat)
    test.append(test_acc)  # Store test accuracy
    print(f"Iteration {i}, train: {train_acc}, test: {test_acc}")
```

```python
plt.plot(range(1, 21), train, label='Train Accuracy')
plt.plot(range(1, 21), test, label='Test Accuracy')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.title('Train vs Test Accuracy over Iterations')
plt.legend()
plt.show()
```
